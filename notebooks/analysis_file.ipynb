```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
df = pd.read_csv('house_price.csv')
```

```python
def show_basic_info(df):
    print("="*50)
    df.info()
    
    print(df.describe())
    
    print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
    print("="*50)
```

```python
def handle_duplicates(df):
    duplicates_count = df.duplicated().sum()
    print(duplicates_count)
    print("="*50)
```

```python
def analyze_categorical_cols(df):
    cat_cols = df.select_dtypes(include='object')
    
    for col in cat_cols.columns:
        unique_vals = cat_cols[col].unique()
        print(f"Column ({col}):")
        print(f"Unique Values: {unique_vals}")
        print(f"Count of Unique Values: {len(unique_vals)}")
        print("-" * 30)
    print("="*50)
```

```python
def plot_numerical_distributions(df):
    nums_cols = df.select_dtypes(exclude='object')
    
    for col in nums_cols.columns:
        fig, axes = plt.subplots(1, 2, figsize=(12, 4))
        
        # Histogram
        sns.histplot(df[col], bins=50, ax=axes[0])
        axes[0].set_title(f'Histogram of {col}')
        
        # KDE
        sns.kdeplot(df[col], fill=True, ax=axes[1])
        axes[1].set_title(f'KDE of {col}')
        
        plt.tight_layout()
        plt.show()
```

```python
def plot_numerical_outliers(df):
    nums_cols = df.select_dtypes(exclude='object')
    
    for col in nums_cols.columns:
        fig, axes = plt.subplots(1, 2, figsize=(12, 4))
        
        # Boxplot
        sns.boxplot(x=df[col], ax=axes[0])
        axes[0].set_title(f'Boxplot of {col}')
        
        # Violinplot
        sns.violinplot(x=df[col], fill=True, ax=axes[1])
        axes[1].set_title(f'Violinplot of {col}')
        
        plt.tight_layout()
        plt.show()
```

```python
def plot_specific_features(df):
    
    nums_cols = df.select_dtypes(exclude='object')
    
    if 'balcony' in nums_cols.columns:
        plt.figure(figsize=(6, 4))
        sns.countplot(x=df['balcony'])
        plt.title('Count of Balcony')
        plt.show()
        
    if 'bath' in nums_cols.columns:
        plt.figure(figsize=(6, 4))
        sns.countplot(x=df['bath'])
        plt.title('Count of Bath')
        plt.show()

    if 'price' in df.columns:
        fig, axes = plt.subplots(1, 2, figsize=(12, 4))
        sns.histplot(df['price'], bins=50, ax=axes[0])
        axes[0].set_title('Histogram of Price')
        
        sns.boxplot(x=df['price'], ax=axes[1])
        axes[1].set_title('Boxplot of Price')
        
        plt.tight_layout()
        plt.show()
```

```python
def run_full_data_analysis(df):
    show_basic_info(df)
    handle_duplicates(df)
    analyze_categorical_cols(df)
    plot_numerical_distributions(df)
    plot_numerical_outliers(df)
    plot_specific_features(df)
```

```python
run_full_data_analysis(df)
```
